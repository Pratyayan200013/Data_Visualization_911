```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame=pd.read_csv('911.csv')
```

```python
dataFrame.info()
```

```python
dataFrame.head()#only gives the head of the data set instead of the full dataset 
```

```python
#top 5 zip codes calling to 911
dataFrame['zip'].value_counts().head(5)
```

```python
#top 10 twp(townships) calling to 911
dataFrame['twp'].value_counts().head(10)
```

```python
#unique reasons to call 911
dataFrame['title'].nunique()
```

```python
dataFrame.head()
```

```python
#specific reasons to call 911
dataFrame['SpecificReasons']=dataFrame['title'].apply(lambda title: title.split(':')[1])
```

```python
dataFrame.head(10)
```

```python
dataFrame['SpecificReasons'].value_counts().head(15)
```

```python
#Broader Reasons to call to 911
dataFrame['BroaderReasons']=dataFrame['title'].apply(lambda x: x.split(':')[0])
```

```python
dataFrame
```

```python
dataFrame['BroaderReasons'].value_counts()
```

```python
#plot countplot for BroaderReasons
sns.countplot(x='BroaderReasons',data=dataFrame)
```

```python
#converting timeStamp from object to actual time stamp
dataFrame['timeStamp']=pd.to_datetime(dataFrame['timeStamp'])
```

```python
dataFrame.head()
```

```python
dataFrame.info()
```

```python
type(dataFrame['timeStamp'].iloc[0])
```

```python
#creating 3 columns for day,month and hours
dataFrame['Hour']=dataFrame['timeStamp'].apply(lambda time: time.hour)
```

```python
dataFrame.head()
```

```python
dataFrame['Day']=dataFrame['timeStamp'].apply(lambda date:date.dayofweek)
dataFrame['Month']=dataFrame['timeStamp'].apply(lambda date:date.month)
```

```python
dataFrame.head()
```

```python
dataFrame
```

```python
dataFrame['Day'].nunique()
```

```python
dataFrame['Month'].nunique()
```

```python
dataFrame['Hour'].nunique()
```

```python
# plot a graph for Months
sns.countplot(x='Month',data=dataFrame,hue='BroaderReasons')
plt.legend(loc='upper right',bbox_to_anchor=(1,0,0.3,1))
```

```python
byDay=dataFrame.groupby('Day').count()#The data here is grouped with respect to Day and the values are retruned according to the function we used (here it is .count())
```
